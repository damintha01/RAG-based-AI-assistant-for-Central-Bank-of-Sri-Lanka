```python
from pypdf import PdfReader

reader = PdfReader(r"D:\AI\GenAI\my_projects\RAG-based-AI-assistant-for-Central-Bank-of-Sri-Lanka\data\raw\Annual Reports\AR_2020_presentation_e.pdf")

text = ""

for page in reader.pages:
    text += page.extract_text() + "\n"

print(text[:1000])
```

```python
import re

def clean_text(text):
    text = re.sub(r'\n+', '\n', text)
    text = re.sub(r'\s+', ' ', text)
    return text.strip()

cleaned_text = clean_text(text)

print(cleaned_text[:1000])
```

```python
from dotenv import load_dotenv
import os

load_dotenv()

PINECONE_API_KEY = os.getenv("PINECONE_API_KEY")

```

```python
from pinecone import Pinecone, ServerlessSpec


pc=Pinecone(api_key=PINECONE_API_KEY)

index_name="rag-p1"

# Check if index exists before creating
if index_name not in pc.list_indexes().names():
    pc.create_index(name=index_name,
                    dimension=384,
                    metric="cosine",
                    spec=ServerlessSpec(cloud="aws",
                                        region="us-east-1"))
    print(f"Index '{index_name}' created. Waiting for it to be ready...")
    
    # Wait for the index to be ready
    import time
    while not pc.describe_index(index_name).status['ready']:
        time.sleep(1)
    print(f"Index '{index_name}' is now ready!")
else:
    print(f"Index '{index_name}' already exists")
```
